```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
import os
```

```python
#obtain list of dates
directory = "Boston Celtics"
list_dates = []
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        list_dates.append(filename)
```

```python
# create dictionary 
final_data = {}
```

```python
opponents = {}
# reads in each csv
for i in list_dates:
    data = pd.read_csv(f"Boston Celtics/{i}", index_col=0)
    #gathers data and adds to dict
    date = data["Date"].tolist()[0]
    
    opponents[date] = data["Opponent"][0]
    
    data[date] = data["Score"]
    final_data[date] = data[date]
    final_data[date].append
```

```python
# convert dict to dataframe
data_df = pd.DataFrame(final_data)
```

```python
opp = pd.DataFrame.transpose(pd.DataFrame.from_dict(opponents, orient="index"))
```

```python
data_df.append(opp)
```

```python
#create dict
final_result = {}
```

```python
eff = pd.read_csv("Team_defensive_efficency.csv")
eff.head()
```

```python
efficiency = {}

for name in opponents:
    team = opponents[name]
    team_name = team.split(" ")
    
    team_opp = team_name[0]
    
    if team_name[0] == ("Golden" or "New"):
        team_opp = f"{team_name[0]} {team_name[1]}"
    
    if team_name[0] == "Los":
        team_opp = f"LA {team_name[2]}"

    
    efficiency[name] = eff.loc[eff["Unnamed: 0"] == team_opp]["2021"].values[0]
```

```python
eff_df = pd.DataFrame.transpose(pd.DataFrame.from_dict(efficiency, orient="index"))
eff_df.head()
```

```python
data_df = data_df.append(eff_df)
```

```python
#iterate through index
for name in data_df.index:
    #create linear regression model 
    model = LinearRegression()
    
    # get dataframe with each name 
    finder = data_df[data_df.index == name]
    
    #obtain length of dataset
    X = []
    for x in range(len(finder.columns)):
        X.append(x)
        
    #get stats for each player    
    stats = data_df[data_df.index == name].values
    
    # create dataframe for each player
    XX = pd.DataFrame(X)
    XX["Score"] = stats[0]
    length = len(XX[0])
    XX.dropna(inplace=True)
    
    #reshape values
    X_values = XX[0].values.reshape(-1, 1)
    
    #fit linear regression model
    model.fit(X_values, XX["Score"])
    #predict next game
    predict = [length]
    
    #create dataframe of final prediction and add to dict
    prediction = pd.DataFrame(predict)
    prediction[0].values.reshape(-1,1)
    final_prediction = model.predict(prediction[0].values.reshape(-1,1))
    final_result[name] = final_prediction[0]
```

```python
# convert dict to dataframe
Data = pd.DataFrame.from_dict(final_result, orient='index')
```
