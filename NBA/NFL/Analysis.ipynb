```python
years = ["2021", "2020", "2019", "2018", "2017"]
```

```python
players = {}
new_columns = []
for year in years:
    Data = pd.read_csv(f"{year}/Field Goals.csv")
    length = len(Data.columns)
    column = Data.columns[1:length]
    for row, x in Data.iterrows():
        players[x["Unnamed: 0"]] = []
```

```python
for year in years:
    Data = pd.read_csv(f"{year}/Field Goals.csv")
    length = len(Data.columns)
    column = Data.columns[1:length]
    for name in column:
        new_columns.append(f"{year}-{name}")
    testing = []
    for row, x in Data.iterrows():
        testing.append(x["Unnamed: 0"])
        for t in column:
            players[x["Unnamed: 0"]].append(x[t])
    for player in players:
        if player not in testing:
            for t in column:
                players[player].append(0)
```
