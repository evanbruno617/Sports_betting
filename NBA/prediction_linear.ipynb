```python
# create dictionary 
final_data = {}
```

```python
#obtain list of dates
directory = f'{team}'
list_dates = []
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        list_dates.append(filename)

# reads in each csv
for i in list_dates:
    data = pd.read_csv(f"{directory}/{i}", index_col=0)

    #gathers data and adds to dict
    date = data["Date"].tolist()[0]
    data[date] = data["Score"]
    final_data[date] = data[date]

# convert dict to dataframe
data_df = pd.DataFrame(final_data)

t = data_df.transpose()

t['Datetime'] = pd.to_datetime(t.index)

data = t.sort_values(by='Datetime')

data_df = t.sort_values(by='Datetime').transpose()
```

```python
#create dict
final_result = {}
```

```python
#iterate through index
for name in data_df.index[:-1]:
    #create linear regression model 
    model = LinearRegression()
    
    final_result[name] = {}
    
    # get dataframe with each name 
    finder = data_df[data_df.index == name]
    
    #obtain length of dataset
    X = []
    for x in range(len(finder.columns)):
        X.append(x)
        
    stats = data[[name]].dropna().head(3).transpose()
            
    #obtain length of dataset
    X = []
    for x in range(len(stats.columns)):
        X.append(x)

    # create dataframe for each player
    XX = pd.DataFrame(X)
    XX["Score"] = stats[0]
    XX.dropna(inplace=True)
    length = len(XX[0])
    
    #reshape values
    X_values = XX[0].values.reshape(-1, 1)
    
    #fit linear regression model
    model.fit(X_values, XX["Score"])
    #predict next game
    predict = [length]
    
    #create dataframe of final prediction and add to dict
    prediction = pd.DataFrame(predict)
    prediction[0].values.reshape(-1,1)
    final_prediction = model.predict(prediction[0].values.reshape(-1,1))
    final_result[name]['Predict'] = final_prediction[0]
        
    final_result[name]['R^2'] = r2_score(XX["Score"], model.predict(X_values))
```

```python
# convert dict to dataframe
Data = pd.DataFrame.from_dict(final_result, orient='index')
```

```python
teams = [
    "Atlanta Hawks",
    "Boston Celtics",
    "Brooklyn Nets",
    "Charlotte Hornets",
    "Chicago Bulls",
    "Cleveland Cavaliers",
    "Dallas Mavericks",
    "Denver Nuggets",
    "Detroit Pistons",
    "Golden State Warriors",
    "Houston Rockets",
    "Indiana Pacers",
    "LA Clippers",
    "Los Angeles Lakers",
    "Memphis Grizzlies",
    "Miami Heat",
    "Milwaukee Bucks",
    "Minnesota Timberwolves",
    "New Orleans Pelicans",
    "New York Knicks",
    "Oklahoma City Thunder",
    "Orlando Magic",
    "Philadelphia 76ers",
    "Phoenix Suns",
    "Portland Trail Blazers",
    "Sacramento Kings",
    "San Antonio Spurs",
    "Toronto Raptors",
    "Utah Jazz",
    "Washington Wizards"
]
```

```python
# test

nums = [3, 5, 7, 10]

results = {}

for num in nums:

    #create dict
    final_result = {}

    for team in teams:

        #obtain list of dates
        directory = f'Historical/{team}'
        list_dates = []
        for filename in os.listdir(directory):
            if filename.endswith(".csv"):
                list_dates.append(filename)

        # reads in each csv
        for i in list_dates:
            data = pd.read_csv(f"{directory}/{i}", index_col=0)

            #gathers data and adds to dict
            date = data["Date"].tolist()[0]
            data[date] = data["Score"]
            final_data[date] = data[date]

        # convert dict to dataframe
        data_df = pd.DataFrame(final_data)

        t = data_df.transpose()

        t['Datetime'] = pd.to_datetime(t.index)
        
        data = t.sort_values(by='Datetime')

        data_df = t.sort_values(by='Datetime').transpose()

        #iterate through index
        for name in data_df.index[:-1]:
            #create linear regression model 
            model = LinearRegression()

            final_result[name] = {}

            # get dataframe with each name 
            finder = data_df[data_df.index == name]

            #obtain length of dataset
            X = []
            for x in range(len(finder.columns)):
                X.append(x)

            stats = data[[name]].dropna().head(num).transpose()
            
            #obtain length of dataset
            X = []
            for x in range(len(stats.columns)):
                X.append(x)

            stats = stats[stats.index == name].values

            # create dataframe for each player
            XX = pd.DataFrame(X)
            XX["Score"] = stats[0]
            XX.dropna(inplace=True)
            length = len(XX[0])
            
            if length < num:
                continue

            #reshape values
            X_values = XX[0].values.reshape(-1, 1)

            #fit linear regression model
            model.fit(X_values, XX["Score"])
            #predict next game
            predict = [length]

            #create dataframe of final prediction and add to dict
            prediction = pd.DataFrame(predict)
            prediction[0].values.reshape(-1,1)
            final_prediction = model.predict(prediction[0].values.reshape(-1,1))
            final_result[name]['Predict'] = final_prediction[0]

            final_result[name]['R^2'] = r2_score(XX["Score"], model.predict(X_values))
            
    # convert dict to dataframe
    Data = pd.DataFrame.from_dict(final_result, orient='index')
    
    m = pd.DataFrame(Data['R^2'].tolist()).dropna()[0].mean()
    
    results[num] = m
```
