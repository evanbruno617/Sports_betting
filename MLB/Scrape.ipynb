```python
for iteration in range(5):

    html = browser.html
    page = soup(html, 'html.parser')

    href = page.find(class_="button2 prev").get("href")
    
    url = f"https://www.baseball-reference.com{href}"

    browser.visit(url)

    date = page.find(class_="button2 current").get_text()
    
    print(date)

    href_list = []
    for i in page.find_all(class_="right gamelink"):
        t = i.find("a").get("href")
        href_list.append(t)
    myset = set(href_list)
    href_links = list(myset)

    for link in href_links:
        browser.visit(f"https://www.baseball-reference.com{link}")
        time.sleep(3)
        html = browser.html
        test = soup(html, 'html.parser')
        for x in test.find_all(class_="table_container is_setup")[:4]:
            tablename = x.get("id")[4:]
            table = x.find("table").find("tbody").find_all("tr")
            column_name = []
            for i in x.find("thead").find_all("th")[1:]:
                column_name.append(i.get_text())
            stats = {}
            for i in table:
                name = i.find("th").get_text()
                stats[name] = []
                info = i.find_all("td")
                for t in info:
                    stats[name].append(t.get_text())
                data = pd.DataFrame.from_dict(stats, orient="index", columns=column_name)

                details = test.find_all(id="div_8881013116")[0].find_all("div")[-1].get_text()[20:].split(",")

            if len(details) == 3:    
                data["Temp"] = details[0]
                data["Wind"] = details[1]
                data["Arena"] = details[2]
            else:
                data["Temp"] = details[0]
                data["Wind"] = details[1]
                data["Weather"] = details[2]
                data["Precipitation"] = details[3]

            mypath = tablename
            if not os.path.isdir(mypath):
                os.makedirs(mypath)
                    
            print(f"Table: {date}")

            data.to_csv(f"{mypath}/{date}.csv", index=True)
                
    browser.visit(url)
    
    time.sleep(5)
```
